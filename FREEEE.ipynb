```python
import pandas as pd
data = pd.read_csv('melb_house.csv')
```

```python
subset_area = ['Abbotsford']
subset_data = data[data['suburb'].isin(subset_area)]
subset_data
```

```python
average_cost_subset = subset_data.groupby('suburb')['price'].mean()
average_cost_subset.head()
```
